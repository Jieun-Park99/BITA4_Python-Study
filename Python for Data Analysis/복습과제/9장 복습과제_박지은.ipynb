```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)
axes[0,0].plot(time_hrs, web_monday, color='blue')
axes[0,1].plot(time_hrs, web_tuesday, color='blue')
axes[1,0].plot(time_hrs, web_wednesday, color='blue')
axes[1,1].plot(time_hrs, web_tursday, color='blue')
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
fig, ax = plt.subplots(sharex=True, sharey=True)
ax.plot(time_hrs,web_monday, color='black', label ='Mon', linestyle='--', marker='o')
ax.plot(time_hrs,web_tuesday, color='red', label='Tues', linestyle='--', marker='o')
ax.plot(time_hrs,web_wednesday, color='blue', label='Wed', linestyle='--', marker='o')
ax.plot(time_hrs,web_tursday, color='green', label='Turs', linestyle='--', marker='o')
ax.legend()
ax.set_xlabel('Hrs')
ax.set_ylabel('Number of users')
ax.set_title('Web site traffic')
ymax = max(max(web_monday), max(web_tuesday), max(web_wednesday), max(web_tursday))
for dat in (web_monday,web_tuesday,web_wednesday, web_tursday):
    if ymax in dat:
        xpos = dat.index(ymax)
xmax = time_hrs[xpos]
ax.annotate('Max', xy=(xmax, ymax+50), xytext=(xmax+2, ymax+40),
            arrowprops=dict(facecolor='green', headwidth=8, width=4,
                           headlength=10),
            horizontalalignment='left',verticalalignment='top'
            )
```

```python
import pandas as pd
%matplotlib inline
data = pd.read_csv('train.csv', encoding='utf-8')
```

## 3_2

```python
dd = data.iloc[:,[1,8,9,10,11,12,13]]
dd
```

```python
dd["date"] = dd["date"].astype('datetime64[ns]')
dd['date']
```

```python
import datetime
d =dd['date'].dt.weekday
d
dd['date']=d
```

```python
def mean(df):
    return df.mean()
```

```python
new = dd.groupby(['date']).apply(mean)
new.rename(columns={'date':'weekday'},inplace=True)
new
```

```python
import seaborn as sns
```

```python
figure, ((ax1,ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(nrows=2, ncols=3)
figure.set_size_inches(18,8)
sns.barplot(x=new['weekday'], y=new.iloc[:,1], ax=ax1)
sns.barplot(x=new['weekday'], y=new.iloc[:,2], ax=ax2)
sns.barplot(x=new['weekday'], y=new.iloc[:,3], ax=ax3)
sns.barplot(x=new['weekday'], y=new.iloc[:,4], ax=ax4)
sns.barplot(x=new['weekday'], y=new.iloc[:,5], ax=ax5)
sns.barplot(x=new['weekday'], y=new.iloc[:,6], ax=ax6)
```
