## 1. Feb 15, 2019 04:30PM을 a에 이대로 입력하여 변수를 만드시오
## 같은 방법으로 Feb 15, 2020 04:30PM을 b에 입력하시오
## 둘다 형태는 Timestamp가 되도록 만드시오

```python
import pandas as pd
import numpy as np
from dateutil.parser import parse
```

```python
a = pd.to_datetime(parse('Feb 15, 2019 04:30PM'))
b  = pd.to_datetime(parse('Feb 15, 2020 04:30PM'))
```

## 2. a부터 b까지 월 마지막날, 시각은 자정에 맞춰서 dats를 만드시오. 
## (dtype은 datetime[ns]가 나와야 합니다)

```python
dats = pd.date_range(a,b,freq='BM', normalize=True)
dats
```

## 3. 같은 기간으로 빈도를 매월 2째주 월요일로 불러오시오

```python
dats = pd.date_range(a,b,freq='WOM-2MON')
```

```python
list(dats)
```

## 4. timestamp를 index로 갖는 데이터프레임 ts를 생성하시오(1000개의 난수 생성, 5개의 열 생성)
## 열이름은 A,B,C,D,E
## a부터 기간 시작(위에 설정했던)

```python
ts = pd.DataFrame(np.random.randn(1000,5),
              index=pd.date_range(a,periods=1000, freq='M'),
                 columns = ['A','B','C','D','E'])
ts
```

## 4-1 평균을 구하며 downsampling하시오(기간은 본인이 알아서 설정)

```python
ts.resample('2M').mean()
```

## 4-2. 뒤에서 나올 데이터를 쓰며 upsampling하시오 (기간은 본인이 설정)

```python
ts.resample('15D').bfill()
```

## 5. index를 월단위 기간, 그리고 데이터는 평균으로 리샘플하여 tsm에 저장하시오

```python
tsm = ts.resample('M',kind='period').mean()
tsm
```

## 5-1. 데이터의 index가 unique한지 확인하시오
## 그리고 new_tsm에 연별로 묶어 평균을 구하시오 단, na값은 0으로 채우시오

```python
tsm.index.is_unique
```

```python
new_tsm = tsm.resample('Y', how=np.mean).fillna(0)
new_tsm
```

## 5-2. new_tsm 데이터를 12월을 연도마감으로하는 분기 주기를 계산하시오

```python
new_tsm.resample('Q-DEC').ffill()
```
