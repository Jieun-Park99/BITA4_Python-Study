```python
#2번
ab=pd.read_csv('AB_NYC_2019.csv')
ab1=ab.iloc[:,-8:]
ab1.head()
ab2=ab1.dropna()
ab2.head()
```

```python
#2-2
ab2['minimum_monthly_sales']=ab2['reviews_per_month']*ab2['minimum_nights']*ab2['price']
ab2.head()
```

```python
#2-3-1
def top(df,n=10,column='minimum_monthly_sales'):
    return df.sort_values(by=column)[-n:]
top(ab2,n=10)
```

```python
#2-3-2
ab2.groupby('room_type').apply(top)
```

```python
#2-4
ab2.pivot_table(['minimum_monthly_sales'],index=['room_type'])
```
