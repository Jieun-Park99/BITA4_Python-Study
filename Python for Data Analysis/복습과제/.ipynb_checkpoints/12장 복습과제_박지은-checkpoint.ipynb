## Q1-2

```python
df3 = pd.DataFrame({'date': times.repeat(3),
                   'key': np.tile(['a','b','c'],N),
                   'value':np.arange(N*3.)})
df3
```

```python
time_key = pd.Grouper(freq="3H")
```

```python
resampled = (df3.set_index('date')
             .groupby(['key',time_key])
             .sum())
resampled
```

## Q2

```python
df4 = pd.DataFrame({'name':['은조','재훈','광민','정민','지은','지선','봉석','세정','석희'],
                   'height':[163,176,183,177,172,159,175,162,173]})
df4
```

```python
def StandWeight(height):
    weight=[]
    for h in height:
        if h < 150:
            w = h-100
        elif h >= 150:
            w = (h-100)*0.9
        weight.append(w)
    return pd.Series(weight)
```

```python
StandWeight(df4.height)
```

## Q3

```python
np.random.seed(9195)
draws = np.random.randint(1,100,size=20)
bins = pd.qcut(draws,4,labels=['Q1','Q2','Q3','Q4'])
bins = pd.Series(bins, name='Q')

results = (pd.Series(draws)
          .groupby(bins)
          .agg(['count','mean']))
results
```
