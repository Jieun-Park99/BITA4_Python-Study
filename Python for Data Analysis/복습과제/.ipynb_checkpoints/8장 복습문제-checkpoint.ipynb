### 2-1. 4x3 크기의 데이터프레임을 랜덤으로 만드세요

```python
import numpy as np
```

```python
df1 = pd.DataFrame(np.random.randn(4,3),columns=['a','b','c'])
df2 = pd.DataFrame(np.random.randn(3,2),columns=['d','c'])
```

```python
df1
```

```python
df2
```

### 2-2. df1과 df2를 이어 붙인 후 겹치는 데이터를 합치세요

```python
link_d = pd.merge(df1,df2,how='outer')
link_d
```

```python
combine_d = df1.combine_first(df2)
combine_d
```

### 3-1. example 폴더에서 macrodata.csv를 data라는 이름으로 로드해주세요

```python
data = pd.read_csv('../examples/macrodata.csv',encoding='euc-kr')
data
```

### 3-2. pivot함수를 활용하여 다음과 같이 연도별 분기별 realgdp값을 출력하시오

```python
yq_realgdp = data.pivot('year','quarter','realgdp')
yq_realgdp.head()
```

### 3-3. pivot함수를 활용하여 다음과 같이 연도별 분기별 cpi값을 출력하세요

```python
yq_cpi = data.pivot('year','quarter','cpi')
yq_cpi.head()
```
