# 보너스 문제

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'key1' : ['a','a','a','b','b','b','aa','aa','aa'], 
                   'key2' : ['one','two','one','two','two','one','one','two','two'], 
                   'data1' : np.random.randn(9), 
                   'data2' : np.random.randn(9)})
df
```

# 보너스-1 

```python
df.groupby(['key2','key1'])['data2'].mean().unstack()
```

# 1

```python
dat = pd.read_csv('AB_NYC_2019.csv').iloc[:,-8:]
dat.dropna(inplace=True)
dat
```

# 2

```python
dat['minimum_monthly_sales'] = dat['reviews_per_month']*dat['minimum_nights']*dat['price']
dat
```

# 3-1

```python
def top(df, n=10, column='minimum_monthly_sales'):
    return df.sort_values(by=column)[-n:]
top(dat)
```

# 3-2

```python
dat.groupby('room_type').apply(top)
```

# 4

```python
dat.pivot_table(['minimum_monthly_sales'], index=['room_type'])
```
